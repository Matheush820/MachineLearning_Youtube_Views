```python
#  Importação de Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de exibição
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')

print("Bibliotecas importadas com sucesso!")
```

```python
# 2️⃣ Carregamento dos Dados
df = pd.read_csv('./youtube_views.csv')

# Exibir informações iniciais
print("Formato do dataset:", df.shape)
display(df.head())
display(df.tail())
```

```python
# Visão Geral do Dataset
print("Tipos de dados:")
display(df.dtypes)

print("\nResumo estatístico (variáveis numéricas):")
display(df.describe())

# Separação por tipo
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(exclude=['int64', 'float64']).columns

print(f"\nVariáveis numéricas ({len(num_cols)}): {list(num_cols)}")
print(f"Variáveis categóricas ({len(cat_cols)}): {list(cat_cols)}")
```

```python
# Análise de Valores Faltantes
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100

print("Valores faltantes por coluna:")
display(pd.DataFrame({'Total': missing, '%': missing_percent.round(2)}))

# Gráfico
plt.figure(figsize=(10, 5))
sns.barplot(x=missing_percent, y=missing_percent.index)
plt.title("Percentual de Valores Faltantes por Coluna")
plt.xlabel("% de valores faltantes")
plt.ylabel("Colunas")
plt.show()
```

```python
# 5️ Análise da Variável Alvo
target_col = 'final_grade'

if target_col not in df.columns:
    # Selecionar automaticamente uma variável numérica semelhante
    possible_targets = [col for col in df.columns if 'view' in col.lower() or 'score' in col.lower() or 'rating' in col.lower()]
    if possible_targets:
        target_col = possible_targets[0]
    else:
        target_col = df.select_dtypes(include=['int64', 'float64']).columns[0]

print(f"Variável alvo selecionada: {target_col}")

# Estatísticas
display(df[target_col].describe())

# Histograma
plt.figure(figsize=(8, 4))
sns.histplot(df[target_col], bins=30, kde=True)
plt.title(f"Distribuição da Variável Alvo: {target_col}")
plt.xlabel(target_col)
plt.ylabel("Frequência")
plt.show()
```

```python
#  Análise Univariada - Numéricas
for col in num_cols:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribuição - {col}")

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot - {col}")
    plt.tight_layout()
    plt.show()

    print(f"Estatísticas de {col}:")
    display(df[col].describe())

    # Outliers (IQR)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"Outliers em {col}: {len(outliers)}\n{'-'*40}")
```

```python
# 7️ Análise Univariada - Categóricas
for col in cat_cols:
    print(f"\nAnálise da variável categórica: {col}")
    print(df[col].value_counts().head(10))
    
    plt.figure(figsize=(8,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index[:10])
    plt.title(f"Distribuição das Categorias - {col}")
    plt.xlabel("Contagem")
    plt.ylabel(col)
    plt.show()
```

```python
# 8️ Correlação
plt.figure(figsize=(12, 8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matriz de Correlação entre Variáveis Numéricas")
plt.show()

# Correlação com a variável alvo
if target_col in num_cols:
    corr_target = corr[target_col].sort_values(ascending=False)
    print("Correlação das variáveis numéricas com o alvo:")
    display(corr_target)
```

```python
# Análise Bivariada
if target_col in df.columns:
    for col in cat_cols:
        if df[col].nunique() <= 10:  # até 10 categorias
            plt.figure(figsize=(8,4))
            sns.boxplot(x=col, y=target_col, data=df)
            plt.title(f"{target_col} por {col}")
            plt.show()
```

```python
# Identificação de Outliers
outlier_summary = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    count = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    outlier_summary[col] = count

outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Qtd_Outliers']).sort_values(by='Qtd_Outliers', ascending=False)
display(outlier_df)

plt.figure(figsize=(10,5))
sns.barplot(x=outlier_df.Qtd_Outliers, y=outlier_df.index)
plt.title("Quantidade de Outliers por Variável")
plt.xlabel("Quantidade")
plt.ylabel("Variável")
plt.show()
```
